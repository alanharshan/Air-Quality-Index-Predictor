## Train-Test Splitting

```python
from sklearn.model_selection import train_test_split
train_set,test_set= train_test_split(dataset,test_size=0.2,random_state=42)
print(f"Train set has {len(train_set)} rows") 
print(f"Test set has {len(test_set)} rows") 
```

```python
train_set["VV"].value_counts()
```

```python
dataset=train_set.copy()  # proceed with train_set
```

## Correlations

```python
corr_matrix=dataset.corr()
```

```python
corr_matrix["PM 2.5"]
```

```python
from pandas.plotting import scatter_matrix
attributes=["PM 2.5","T","SLP","H","VV","V","VM"]
scatter_matrix(dataset[attributes],figsize=(12,8))
```

## Missing Attributes

```python
dataset.info() ## before imputing
```

```python
from sklearn.impute import SimpleImputer
imputer=SimpleImputer(strategy="median")
imputer.fit(dataset)
imputer.statistics_ 
```

```python
X=imputer.transform(dataset)
X
```

```python
dataset_tr=pd.DataFrame(X,columns=dataset.columns)
```

```python
dataset_tr.info()
```

```python
dataset_tr.head()
```

```python
test_set.info()
```

```python
dataset=dataset_tr.drop(["PM 2.5"],axis=1)
dataset_labels=dataset_tr["PM 2.5"].copy()  ##separating labels from dataset
```

```python
dataset.head()
```

## Creating a pipeline

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
my_pipeline=Pipeline([
    ('imputer',SimpleImputer(strategy="median")),
    ('std_scaler',StandardScaler())
]
)
```

```python
dataset_num_tr=my_pipeline.fit_transform(dataset)
```

```python
dataset_num_tr.shape
```

## Selecting a desired model

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
model=RandomForestRegressor()
model.fit(dataset_num_tr,dataset_labels)
```

```python
some_data=dataset.iloc[:5]
```

```python
some_labels=dataset_labels.iloc[:5]
```

```python
prepared_data=my_pipeline.transform(some_data)
```

```python
model.predict(prepared_data)
```

```python
list(some_labels)
```

## Evaluating the model

```python
import numpy as np
from sklearn.metrics import mean_squared_error
dataset_prediction=model.predict(dataset_num_tr)
mse=mean_squared_error(dataset_labels,dataset_prediction)
rmse=np.sqrt(mse)
```

```python
rmse     
```

```python
from sklearn.model_selection import cross_val_score
scores=cross_val_score(RandomForestRegressor(),dataset_num_tr,dataset_labels,scoring="neg_mean_squared_error",cv=10)
rmse_scores=np.sqrt(-scores)
```

```python
rmse_scores
```

```python
def print_scores(scores):
    print("scores:",scores)
    print("mean:",scores.mean())
    print("std:",scores.std())
```

```python
print_scores(rmse_scores)
```

## Saving the model

```python
from joblib import dump, load
dump(model, 'Delhi_Air_Quality.joblib') 
```

## Testing the model

```python
X_test=test_set.drop(["PM 2.5"],axis=1).copy()
Y_test=test_set["PM 2.5"].copy()
X_test_prepared=my_pipeline.transform(X_test)
final_predictions=model.predict(X_test_prepared)
final_mse=mean_squared_error(Y_test,final_predictions)
final_rmse=np.sqrt(final_mse)
```
